```python
import os
import time

import cv2 
import torch
import numpy as np
import torchvision
import matplotlib.pyplot as plt
import cheapfake.contrib.dataset as dataset
import cheapfake.contrib.video_processor as video_processor
```

Sample frames and audio stochastically. Frames are extracted non-sequentially and the audio is extracted sequentially.

```python
random_seed = 30
root_path = "/Users/shu/Documents/Datasets/DFDC_small_subset_raw"

dfdataset = dataset.DeepFakeDataset(root_path=root_path, random_seed=random_seed, return_tensor=False, sequential_frames=False, sequential_audio=True, stochastic=True)

start_time = time.time()
frames, audio = dfdataset.__getitem__(0)
end_time = time.time()

print("The entire operation took {} seconds".format(end_time - start_time))
```

```python
# Plot the frames.
frames_plotting = np.einsum("ijkl->iklj", frames)

fig = plt.figure(figsize=(10, 7))

plt.subplot(1, 3, 1)
plt.imshow(frames_plotting[0].astype("uint8"))

plt.subplot(1, 3, 2)
plt.imshow(frames_plotting[1].astype("uint8"))

plt.subplot(1, 3, 3)
plt.imshow(frames_plotting[2].astype("uint8"))
```

```python
# Plot the audio.
fig = plt.figure(figsize=(10, 7))
plt.plot(audio)
```
