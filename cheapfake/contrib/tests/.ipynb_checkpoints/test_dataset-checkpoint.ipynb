```python
import os
import time

import cv2 
import torch
import numpy as np
import torchvision
import matplotlib.pyplot as plt
import cheapfake.contrib.dataset as dataset
import cheapfake.contrib.video_processor as video_processor
```

Sample frames and audio stochastically. Frames are extracted non-sequentially and the audio is extracted sequentially.

```python
random_seed = 8
root_path = "/Users/shu/Documents/Datasets/DFDC_small_subset_raw"

dfdataset = dataset.DeepFakeDataset(root_path=root_path, random_seed=random_seed, return_tensor=False, sequential_frames=False, sequential_audio=True, stochastic=True)

start_time = time.time()
frames, audio, _ = dfdataset.__getitem__(0)
end_time = time.time()

print("The entire operation took {} seconds".format(end_time - start_time))
print(frames.shape)
print(audio.shape)
```

```python
# Plot the frames.
frames_plotting = np.einsum("ijkl->jkli", frames)

fig = plt.figure(figsize=(10, 7))

plt.subplot(1, 3, 1)
plt.imshow(frames_plotting[0].astype("uint8"))

plt.subplot(1, 3, 2)
plt.imshow(frames_plotting[1].astype("uint8"))

plt.subplot(1, 3, 3)
plt.imshow(frames_plotting[2].astype("uint8"))
```

```python
# Plot the audio.
fig = plt.figure(figsize=(10, 7))
plt.plot(audio)
```

```python
# Ensure that the two audio samples actually differ between two random seeds.
root_path = "/Users/shu/Documents/Datasets/DFDC_small_subset_raw"

random_seed_0 = 41
dfdataset_0 = dataset.DeepFakeDataset(root_path=root_path, sequential_audio=True, return_tensor=False, sequential_frames=False, stochastic=True, random_seed=random_seed_0)
frames_0, audio_0, _ = dfdataset_0.__getitem__(0)

random_seed_1 = 30
dfdataset_1 = dataset.DeepFakeDataset(root_path=root_path, sequential_audio=True, return_tensor=False, sequential_frames=False, stochastic=True, random_seed=random_seed_1)
frames_1, audio_1, _ = dfdataset_1.__getitem__(0)

diff = np.linalg.norm(audio_0 - audio_1)
print(diff)
```

```python
fig = plt.figure(figsize=(20, 7))

plt.subplot(1, 2, 1)
plt.plot(audio_0)

plt.subplot(1, 2, 2)
plt.plot(audio_1)
```
