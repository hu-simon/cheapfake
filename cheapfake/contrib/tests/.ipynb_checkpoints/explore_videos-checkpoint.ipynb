Creates a new dataframe containing the file locations and indices of videos that are horizontally shot.

```python
import os
import time

import cv2
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
metadata_path = "/home/shu/cheapfake/cheapfake/contrib/balanced_metadata_fs03.csv"
dataframe = pd.read_csv(metadata_path)

dataframe.head()
```

```python
filelist = dataframe[["Files", "label"]]
filelist.head()
```

```python
filelist_subset = filelist[:10]

for idx, file in enumerate(filelist_subset["Files"]):
    vidcap = cv2.VideoCapture(file)
    height = vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width = vidcap.get(cv2.CAP_PROP_FRAME_WIDTH)
    print(height, width)
    if height < width:
        print("This is a wide video, at index {} and the label is {}".format(idx, filelist_subset["label"][idx]))
```

```python
vidcap = cv2.VideoCapture(filelist_subset["Files"][7])
success, image = vidcap.read()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(image)
plt.show()
```

Armed with the above information and procedure, we can create a new dataframe that only contains the wide videos.

```python
videos = list()
labels = list()

for idx, file in enumerate(tqdm(filelist["Files"])):
    vidcap = cv2.VideoCapture(file)
    height = vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width = vidcap.get(cv2.CAP_PROP_FRAME_WIDTH)
    if height < width:
        videos.append(file)
        labels.append(filelist["label"][idx])
```

```python
df = pd.DataFrame({"Files": videos, "label": labels})
df.head()
```
